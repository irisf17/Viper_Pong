# High Scores for TTA subjects

```python
# Subject 1
TTA1_scores_1 = np.array([6,6,11,13])
TTA1_rounds_1 = np.linspace(1, len(TTA1_scores_1), len(TTA1_scores_1))

TTA1_scores_2 = np.array([5,18,11])
TTA1_rounds_2 = np.linspace(1, len(TTA1_scores_2), len(TTA1_scores_2))
# Subject 2
TTA2_scores_1 = np.array([1,1,4,10,9])
TTA2_rounds_1 = np.linspace(1, len(TTA2_scores_1), len(TTA2_scores_1))

TTA2_scores_2 = np.array([3,4,2,17,1  ])
TTA2_rounds_2 = np.linspace(1, len(TTA2_scores_2), len(TTA2_scores_2))

# first play
TTA_y_1 = np.array([6, 1,   6,1,  11,4,   13,10,   9])
TTA_x_1 = np.array([1,1,    2,2,  3,3,   4,4,    5])

# last play
TTA_y_2 = np.array([5,3,   18,4,   11,2,    17,   1])
TTA_x_2 = np.array([1,1,    2,2,    3,3,     4,   5])
```

# Plotting for healthy subjects

```python
plt.scatter(x_1, y_1, color='c', alpha=0.7, edgecolors="k")
plt.scatter(x_2, y_2, color = 'm', alpha=0.7, edgecolors="k")

a,b = np.polyfit(x_1, y_1,1)
c,d = np.polyfit(x_2, y_2,1)

xseq_1 = np.linspace(0, max(x_1), num=10)
xseq_2 = np.linspace(0, max(x_2), num=10)


plt.plot(xseq_1, a*xseq_1+b, color='c', linewidth = 2)
plt.plot(xseq_2, c*xseq_2+d, color='m', linewidth = 2)

plt.xlim(0.5, 6.5)
plt.title("High score for healthy subjects, first 15 min vs. last 15 min")
plt.xlabel("Number of round")
plt.ylabel("High score")
plt.legend(['first 15 min', 'last 15 min', 'best fit of line for first 15 min', 'best fit of line for last 15 min']) 
plt.grid()
plt.show() 
```

# Generating heat maps

Healthy subjects first 15 min

```python
# Subject 1
sub1_scores = np.array([11,8,3,4,11])

# Subject 2
sub2_scores = np.array([9,5,9,19,0])

# Subject 3
sub3_scores = np.array([5,12,9,16,0])

# Subject 4
sub4_scores = np.array([5,3,1,11,11])

# Subject 5
sub5_scores = np.array([17,11,10,5,0])



# Create a heatmap for subjects without feedback
data = np.array([sub1_scores, sub2_scores,sub3_scores,sub4_scores,sub5_scores])

# Labels
xlabs = ["1", "2", "3", "4",
         "5"]
ylabs = ["sub:1", "sub:2", "sub:3", "sub:4",
         "sub:5"]

# Heat map
fig, ax = plt.subplots()
im = ax.imshow(data, vmax = 26)

# Add the labels
ax.set_xticks(np.arange(len(xlabs)), labels = xlabs)
ax.set_yticks(np.arange(len(ylabs)), labels = ylabs)

ax.set_title("Healthy subjects highscores, first 15 min")
ax.set_xlabel("round of play")

for i in range(len(xlabs)):
    for j in range(len(ylabs)):
        if data[j,i] != 0:
            text = ax.text(i,j, round(data[j, i], 1),
                           ha = "center", va = "center", color = "k")

# Add the color bar
cbar = ax.figure.colorbar(im, ax = ax)
cbar.ax.set_ylabel("Color bar", rotation = -90, va = "bottom")
```
